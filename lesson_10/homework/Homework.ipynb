In this homework we are comparing two trackers - KCF and CSRT.

Main code with tracking implementation is in homework.py. Below is the comparison of the results and answers for the questions provided.

```python
import os
import cv2
import matplotlib.pyplot as plt

def show_saved_frames(output_folder, tracker_type):
    frame_indices = [0, 4, 9]
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    for i, frame_index in enumerate(frame_indices):
        frame_path = os.path.join(output_folder, f"{tracker_type}_frame_{frame_index}.jpg")
        if os.path.exists(frame_path):
            frame = cv2.imread(frame_path)
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            axes[i].imshow(frame)
            axes[i].set_title(f"{tracker_type} Frame {frame_index}")
            axes[i].axis('off')
        else:
            print(f"Warning: {frame_path} does not exist.")
    
    plt.tight_layout()
    plt.show()

output_folder = "output_frames"
tracker_types = ["KCF", "CSRT"]
```

Compare the results:

* Do you see any differences? If so, what are they?
* Does one tracker perform better than the other? In what way?

A1: CSRT definitely performed better overall. It is consistent with initial bbox, no matter the box size and number of distinctive features inside it, and never loses the tracked object.



```python
show_saved_frames(output_folder, "CSRT")
```

A2: At the same time KCF was less reliable. With smaller boxes and/or less features it often fails the tracking after a few frames. It also has this weird behaviour shifting the initial box coordinates for no apparent reason.

```python
 show_saved_frames(output_folder, "KCF")
```
